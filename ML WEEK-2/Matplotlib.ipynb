1.Create a figure object called fig using plt.figure() , Use add_axes to add an axis to the figure
canvas at [0,0,1,1]. Call this new axis ax.

```python
import matplotlib.pyplot as plt
import pandas as pd
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
ax.plot(x, y)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Plot on Axis')
plt.show()
```

3.Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 3 for value in X]
print("Values of X:")
print(*range(1,50)) 
print("Values of Y (thrice of X):")
print(Y)
# Plot lines and/or markers to the Axes.
plt.plot(X, Y)
# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title 
plt.title('Draw a line.')
# Display the figure.
plt.show()
```

4.Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
with open("matplot.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
plt.plot(x, y)
# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title 
plt.title('Sample graph!')
# Display a figure.
plt.show()
```

5.Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt

# Data for the lines
x_values = [1, 2, 3, 4, 5]
y1_values = [2, 4, 6, 8, 10]
y2_values = [1, 3, 5, 7, 9]

# Plotting the lines with different widths and colors
plt.plot(x_values, y1_values, label='Line 1', color='black', linewidth=2)
plt.plot(x_values, y2_values, label='Line 2', color='pink', linewidth=3)

# Adding legends
plt.legend()

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')

# Display the plot
plt.show()
```

7.Create plot using subplots(), scatter, and some other methods on the Axes object:

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots with 2 rows and 1 column
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Plot on the first subplot
axs[0].plot(x, y1, color='blue', label='Sin(x)')
axs[0].set_title('Plot of Sin(x)')
axs[0].legend()

# Plot on the second subplot
axs[1].plot(x, y2, color='red', label='Cos(x)')
axs[1].set_title('Plot of Cos(x)')
axs[1].legend()

plt.tight_layout()

# Display the plot
plt.show()
```

8.Generate random data for x and y and create a scatter plot using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
num_points = 50
x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 20

plt.scatter(x, y, color='blue', marker='o', label='Random Data')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Random Data')

plt.legend()

plt.show()
```

9.Create a bar chart to compare the sales of different products. You can use fictional data or come up
with your own dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [400, 600, 800, 300, 700]

plt.bar(products, sales, color='skyblue')

plt.xlabel('Products')
plt.ylabel('Sales (units)')
plt.title('Product Sales Comparison')

plt.show()
```

10.Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of
the data.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  
data = np.random.normal(loc=0, scale=1, size=1000)  

plt.hist(data, bins=10, edgecolor='black') 
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.show()

```

11.Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create
your own.

```python
import matplotlib.pyplot as plt

categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Utilities', 'Others']
expenses = [2000, 800, 500, 300, 250, 450]
custom_colors = ['blue', 'green', 'red', 'yellow', 'violet', 'pink']

plt.figure(figsize=(8, 8))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=custom_colors)
plt.title('Expense Distribution in Budget')
plt.show()

```
